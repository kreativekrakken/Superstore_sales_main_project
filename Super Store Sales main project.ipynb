## Hypothesis 1: Technology products have the highest profit margin compared to other product categories.

```python
df
```

```python
tot_prof=df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
tot_prof
```

```python
tot_prof.plot(kind='bar')
plt.title('Total profit by category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show();
```

```python
mean_prof=df.groupby('Category')['Profit'].mean().sort_values(ascending=False)
mean_prof
```

```python
mean_prof.plot(kind='bar')
plt.title('Mean profit by category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show();

#conclusion:- our hypothesis is correct because both by mean profit and overall profit metrics
#technology segment tops the other catgories
```

## Hypothesis 2: The East region has the highest sales compared to other regions.

```python
reg_sales=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
reg_sales
```

```python
reg_sales.plot(kind='bar')
plt.title('Sales by region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show();
#conclusion- our hypothesis is wrong because clearly the 'Central' region tops sales 
```

## Hypothesis 3: Sales are higher during certain months of the year. 

```python
df
```

```python
df['Order Month']= df['Order Date'].dt.month
df
```

```python
sale_dt=df.groupby('Order Month')['Sales'].sum()
sale_dt
```

```python
sale_dt.plot(kind='bar')
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show();
```

## Hypothesis 4: Orders with same-day shipping have the lowest rate of returned products. 

```python
df['Ship Mode'].unique()
```

```python
df.columns
```

```python
# now clearly there is no return column, so instead we can treat negative profits as returns
returns_df=df[df['Profit']<0].groupby('Ship Mode')
returns_by_ship_mode=(returns_df.size()/df.groupby('Ship Mode').size())
perc=returns_by_ship_mode*100
perc
```

```python
perc.plot(kind='line')
plt.title('Returns vs Shipping Type')
plt.xlabel('Shipping Type')
plt.ylabel('Returns')
plt.show();

#conclusion:- Clearly same order has the lowest returns rate on same day shipping
```

## Hypothesis 5: The company's profit is more on weekends than on weekdays.

```python
df['Weekday']=df['Order Date'].dt.weekday
df
```

```python
day_prof=df.groupby('Weekday')['Profit'].sum()
day_prof
```

```python
day_prof.plot(kind='bar')
plt.title('Profit according to days')
plt.xlabel('days')
plt.ylabel('Profit')
plt.show();

#conclusion:- Our hypthesis was wrong as clearly weekdays have more profit than weekends
#Here 0-Monday,6-Sunday
```

# Further Exploration

```python
country_sale=df.groupby('Country')['Profit'].sum().sort_values(ascending=False)
country_sale.head(10)
#so clearly the company should focus most on the United States
```

```python
us_data=df[df['Country']=='United States']
us_data.groupby('Category')['Profit'].sum()
#clearly here also Technology profits are higher
```

```python
us_data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
#so even in technology, the company should focus mostly on the 'Copiers' and 'Phones'
```

```python
us_data['Order Year']=us_data['Order Date'].dt.year
us_data
```

```python
us_line=us_data.groupby('Order Month')['Sales','Profit'].sum()
us_line.plot(kind='bar')
plt.xlabel('Order Month')
plt.show()
us_line;
```

```python
us_footfall=us_data.groupby('Order Month')['Row ID'].nunique()
us_footfall.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Footfall')
plt.show();
#so clearly the footfalls is the cause of the Sales and Profits pattern, thats is higher sales n profit figures in second
#half of the year
```
